### 2. Write a 6x6 times table matrix 
 

**by using numpy broadcasting (10 points)**


```python
a = np.zeros(6).reshape((6, 1))
b = np.ones(6)

mat=a+b
mat
```

**by using Python nested for-loop (10 points).**

```python
nrow=6
ncol=6
mat = []

for j in range(ncol):
    row = []
    for i in range(nrow):
        row.append(1.0)
       
    mat.append(row)
np.array(mat)
```
